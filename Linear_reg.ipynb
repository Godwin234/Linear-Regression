```python
#This task here is to fit a line to the MPG dataset. this is a Simple linear regression task
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
column_names = ['MPG', 'Cylinders', 'displacement', 'Horsepower','Weight',
               'Acceleration', 'Model Year','Origin']
raw_dataset = pd.read_csv("auto-mpg.data", names = column_names,
                          na_values = "?", comment = '\t',
                         sep = " ", skipinitialspace = True)
dataset = raw_dataset.copy()
dataset.tail()
```

```python
terget = dataset.pop("MPG") 
displacement = dataset.pop("displacement") 
horsepower= dataset.pop("Horsepower") 
weight = dataset.pop("Weight") 
acceleration = dataset.pop("Acceleration") 
weight.shape
```

```python
x_bar = np.mean(displacement)
y_bar = np.mean(terget)
x_min_xbar = displacement - x_bar
y_min_ybar = terget - y_bar
beta_1 = np.sum(x_min_xbar*y_min_ybar)/ np.sum((x_min_xbar)**2)
beta_0 = y_bar - beta_1 * x_bar
prediction = ([])
for z in displacement:
    r = beta_0 + (beta_1 * z)
    prediction.append(r)
print("beta zero :", beta_0)
print("beta one :", beta_1)
print("prediction of MPG when using Displacement = ",prediction)
```

```python
plt.scatter(displacement, terget )
plt.plot(displacement,prediction, "r" ,label = "model")
plt.title("Scatter plot of the Data and the Model")
plt.xlabel("prediction")
plt.ylabel("Data")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x, y)
```

```python
model = reg.predict(x)
```

```python
plt.scatter(x, y )
plt.plot(x,model, "r" ,label = "model")
plt.title("Scatter plot of the Data and the Model")
plt.xlabel("prediction")
plt.ylabel("Data")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```
