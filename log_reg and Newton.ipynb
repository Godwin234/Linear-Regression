```python
def true_divide(number, array):
    cont = []
    for i in array:
        cont.append(number/i)
    return cont
```

```python
def Gradient_fbeta(x_train,y_train,beta):
    exp = np.exp(np.dot(x_train, beta))
    exp = [x + 1 for x in exp]
    y_hat = true_divide(1, exp)
    grad = np.dot(np.transpose(x_train),np.subtract(y_train , y_hat))
    return grad
```

```python
def steplength_bolddriver(function_x, gradient_x, train, target,lrt, beta, R_factor):
    mew = lrt * 1.1
    while np.all(function_x(target,train,beta) - function_x(target,train, (beta + (mew * np.negative((gradient_x(train,target,beta)))))))<= 0:
        mew_new = mew * R_factor
        mew = mew_new
    return mew
```

```python
def case_weight(x_train,y_train, beta):
    exp = np.exp(np.dot(x_train, beta))
    exp = [x + 1 for x in exp]
    y_hat = true_divide(1, exp)
    CW = np.multiply(y_hat, [x - 1 for x in y_hat])
    CW =CW.reshape(766)
    CW = np.diag(CW)
    return CW
```

```python
def second_deriv(x_train, y_train, C_weight, beta):
    X = np.transpose(x_train)
    secon_der = np.dot(X, np.dot(C_weight(x_train,y_train, beta), x_train))
    return secon_der
```

```python

def minimize_Newton(loss_function, beta,lnr,epoch, x_train,y_train,x_test,y_test,step_lenght, Rfactor):
    log_loss_cont = []
    count = []
    diff = []
    lenr = lnr
    logloss = log_loss(x_test, y_test,beta)
    for i in range(epoch):
        count.append(i)
        log_loss_cont.append(logloss)
        lenr = step_lenght(function_fbeta, Gradient_fbeta, x_train, y_train,lnr, beta, R_factor)
        grad = Gradient_fbeta(x_train,y_train,beta)
        hess = second_deriv(x_train, y_train, case_weight, beta)
        new_beta = beta - lenr * np.dot(hess, grad)
        logloss = log_loss(x_test, y_test,beta)
        differe = np.subtract(function_fbeta(y_train, x_train,beta) , function_fbeta(y_train, x_train, new_beta))
        diff.append(differe)
        beta = new_beta
        if i%20 ==0:   
            print("iterstion = ", i, "log_loss = ", logloss, "difference = ", differe)
    return beta, count, diff, log_loss_cont
```
